# Static Exploration for ANTWERP

```python
import numpy as np
import matplotlib.pyplot as plt
import glob, re, h5py

from utils.constants import *
from utils.h5utils import load_h5_file
```

```python
rows = len(GET_CITIES)
cols = 2
scale_factor = 10
fig, axs = plt.subplots(rows, cols, figsize=(cols * scale_factor, rows * scale_factor))

for i, city in enumerate(GET_CITIES):
    static_mask_once = load_h5_file(f"{BASE_FOLDER}/{city}/{city}_static.h5")[0]
    axs[i].set_title(f"{city} low-res map (495x436)")
    axs[i].imshow(static_mask_once[:495, :436], cmap="gray", vmin=0, vmax=255)
    base_map = load_h5_file(f"{BASE_FOLDER}/{city}/{city}_map_high_res.h5")
    plt.axis("off")
    axs[i].imshow(base_map[0:4950, 0:4360], cmap="gray", vmin=0, vmax=255)
    axs[i].set_title(f"{city} high-res map (4950x4360)")
plt.show()
```
